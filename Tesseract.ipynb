```python
# Install the necessary libraries
!pip install pytesseract
!pip install PyMuPDF
!sudo apt-get install tesseract-ocr

# Import libraries
import pytesseract
from PIL import Image
import fitz  # PyMuPDF
import os
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Function to extract text from a single PDF
def extract_text_from_pdf(pdf_path):
    # Open the PDF file
    pdf_document = fitz.open(pdf_path)
    total_pages = pdf_document.page_count
    all_text = ""

    # Loop through all the pages
    for page_num in range(total_pages):
        # Get the page
        page = pdf_document.load_page(page_num)

        # Render the page to an image
        pix = page.get_pixmap()
        image_path = f"page_image_{page_num}.png"
        pix.save(image_path)

        # Open the image
        image = Image.open(image_path)

        # Extract text using PyTesseract
        text = pytesseract.image_to_string(image)

        # Append the text to the overall text
        all_text += text + "\n"

    return all_text

# Function to process all PDFs in a folder
def extract_text_from_folder(folder_path):
    all_text = ""

    # Loop through all files in the folder
    for filename in os.listdir(folder_path):
        if filename.endswith('.pdf'):
            pdf_path = os.path.join(folder_path, filename)
            text = extract_text_from_pdf(pdf_path)
            all_text += f"Text from {filename}:\n{text}\n"

    return all_text

# Specify the folder path
folder_path = '/content/pdf'  # Change to your folder path

# Extract text from all PDFs in the folder
extracted_text = extract_text_from_folder(folder_path)

# Save the extracted text to a file
with open('extracted_text.txt', 'w') as f:
    f.write(extracted_text)

print(extracted_text)
```
